```python
import pandas as pd
```

```python
data_path = '../data/final_dataset.csv'
```

```python
data = pd.read_csv(data_path)
```

```python
data.drop(data[(data['premium_purchased_next_month'] == 1) & (
    data['premium_purchased_this_month'] == 1)].index, inplace=True)
```

```python
data[(data['premium_purchased_next_month'] == 1) &
     (data['premium_purchased_this_month'] == 1)]
```

```python
grouped_data = data.groupby('user_id')

user_tables = {}
for user_id, group in grouped_data:
    user_tables[user_id] = group.copy()
```

```python
user_df = user_tables[261]
user_df
```

```python
new_table = []
for user_df in user_tables.values():
    condition = user_df['premium_purchased_this_month'] == 1
    filtered_df = user_df[condition]
    try:
        month = filtered_df['month'].iloc[0]
    except Exception:
        month = 12
    new_table.append(user_df[user_df['month'] <= month])
```

```python
merged_data = pd.concat(new_table, ignore_index=True)
```

```python
merged_data
```

```python
merged_data.to_csv("../data/final_data_filtered.csv")
```
