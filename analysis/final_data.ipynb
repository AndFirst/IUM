## Modele

```python
import numpy as np
from sklearn.dummy import DummyClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

### Klasyfikator Dummy

```python
for class_name in TARGET:
    X = data[FEATURES]
    Y = data[class_name]

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    model = DummyClassifier()

    model.fit(X_train, Y_train)

    Y_pred = model.predict(X_test)

    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Skuteczność w przewidywaniu klasy {class_name}: {accuracy:.4f}")

    cm = confusion_matrix(Y_test, Y_pred, labels=[1, 0])
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['Purchased', 'Not purchased'],
                yticklabels=['Purchased', 'Not purchased'])
    plt.xlabel('Predicted')
    plt.ylabel('Real')
    plt.title(f'Confusion Matrix of {class_name}')
    plt.show()
```

Model przewidujący najczęstszą klasę osiąga bardzo wysoką skuteczność dla klas premium_purchased_this_month oraz premium_purchased_next_month. Nie jest to jednak dobra miara ponieważ klasy premium_purchased_this_month oraz premium_purchased_next_month są bardzo niezrównoważone. Zamiast tego użyjemy innej metryki bardziej odpowiedniej dla takich sytuacji. Wydaje sięże ważniejsza będzie minimalizacja przypadków False Negative (lepiej wykryć więcej potencjalnych odbiorców konta premium). Dlatego dobrą miarą wydaje się Recall (Czułość).

### Model drzewiasty

```python
for class_name in TARGET:
    X = data[FEATURES]
    Y = data[class_name]

    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    model = DecisionTreeClassifier()

    model.fit(X_train, Y_train)

    Y_pred = model.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Skuteczność w przewidywaniu klasy {class_name}: {accuracy:.4f}")



```

Prosty model drzewiasty osiąga lepszą skuteczność dla klasy premium_purchased, niż model zwracający zawsze najczęściej występującą klasę. W przypadku dwóch pozostałych model niewiele się nauczył. Może to oznaczać, że nasze dane mają potencjał i po odpowiedniej obróbce i wybraniu lepszego, bardziej zaawansowanego modelu możemy osiągnąć lepsze rezultaty.
